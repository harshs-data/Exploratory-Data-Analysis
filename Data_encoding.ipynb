```python
#Nominal/ OHE
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'status':[ 'single', 'married', 'seperated', 'single', 'single', 'married', 'married']})
```

```python
encoder = OneHotEncoder()
```

```python
encoder
```

```python
encoded = encoder.fit_transform(df[['status']]).toarray()
```

```python
encoder.get_feature_names_out()
```

```python
 encoded_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
 encoded_df
```

```python
encoded_df_dropped = encoded_df.drop(columns = ['status_single'])
encoded_df_dropped
```

```python
pd.concat([df, encoded_df], axis = 1)
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_encoder = OneHotEncoder()
```

```python
tips_encoded = tips_encoder.fit_transform(tips[['sex']]).toarray()
```

```python
tips_encoded_df = pd.DataFrame(tips_encoded, columns = tips_encoder.get_feature_names_out())
```

```python
pd.concat([tips['sex'],tips_encoded_df], axis = 1)
```

```python
t = OneHotEncoder()
```

```python
time_encoder = t.fit_transform(tips[['time']]).toarray()
```

```python
time_encoded = pd.DataFrame(time_encoder, columns = t.get_feature_names_out())
```

```python
pd.concat([tips['time'], time_encoded], axis = 1)
```

```python
# Count values in the 'time' column
counts = tips['time'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Dinner vs Lunch")
plt.show()
```

```python
#Ordinal Encoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({'size': ['small', 'medium', 'large', 'medium', 'small', 'large']})
```

```python
df
```

```python
encoder = OrdinalEncoder(categories= [['small', 'medium', 'large']])
```

```python
encoded = encoder.fit_transform(df[['size']])
encoded_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
```
